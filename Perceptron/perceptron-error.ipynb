```python
# libraries

from random import choice
from numpy import array, dot, random
from pylab import plot, ylim
import matplotlib.pyplot as plt


import neurolab as nl

%matplotlib inline

pl.rcParams['figure.figsize'] = (12.0, 10.0)
```

```python
# funcao de etapa

step = lambda x: 0 if x < 0 else 1
```

```python
# training, input, bias

training_data = [
    (array([0,0,0]), 0),
    (array([1,1,1]), 1),
    (array([1,0,0]), 1),
    (array([0,1,0]), 1),
    (array([0,0,1]), 1),
    (array([1,1,0]), 1),
]
```

```python
# random initial weights

w = random.rand(3)
print(w)
```

```python
# error list, learning rate, learning iteratuions

errors = []
eta = 0.2
n = 50
```

```python
# error magnitudo to 0

for i in range(n):
    x, expected = choice(training_data)
    result = dot(w, x)
    error = expected - step(result)
    errors.append(error)
    w += eta * error * x
```

```python
# learning

for x, _ in training_data:
    result = dot(x, w)
    print("{}: {} -> {}".format(x[:2], result, step(result)))
```
